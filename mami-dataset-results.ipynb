```python
import os
import shutil
import pandas as pd
import json

# Paths
src_images = "/kaggle/input/mami-dataset/MAMI_Dataset/data/MAMI_2022_images"
src_data = "/kaggle/input/mami-dataset/MAMI_Dataset/data"
dst = "/kaggle/working/MIND/data/MAMI"

os.makedirs(dst, exist_ok=True)
os.makedirs(os.path.join(dst, "images"), exist_ok=True)

# Copying all images
for subfolder in ["training_images", "test_images"]:
    folder_path = os.path.join(src_images, subfolder)
    for file in os.listdir(folder_path):
        shutil.copy(os.path.join(folder_path, file), os.path.join(dst, "images"))

# Converting TSV to JSONL
for fname in ["train.tsv", "test.tsv"]:
    tsv_path = os.path.join(src_data, fname)
    if os.path.exists(tsv_path):
        df = pd.read_csv(tsv_path, sep="\t")
        jsonl_path = os.path.join(dst, fname.replace(".tsv", ".jsonl"))
        with open(jsonl_path, "w") as fout:
            for _, row in df.iterrows():
                obj = {
                    "image": row["file_name"],       
                    "label": int(row["label"]),      
                    "text": str(row["text"])
                }
                json.dump(obj, fout)
                fout.write("\n")

base_dirs = {"MAMI": "/kaggle/working/MIND/data/MAMI"}

for dataset, base_path in base_dirs.items():
    for split in ["train.jsonl", "test.jsonl"]:
        in_file = os.path.join(base_path, split)
        if not os.path.exists(in_file):
            continue

        with open(in_file, "r") as fin:
            lines = fin.readlines()
        if split == "test.jsonl":
            lines = lines[:30]
        with open(in_file, "w") as fout:
            fout.writelines(lines)
        print(f" Processed {in_file} ({len(lines)} lines)")
```
